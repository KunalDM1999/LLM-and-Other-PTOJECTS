```python
import requests
import numpy as np
from scipy.spatial.distance import cosine

api_key = "eyJhbGciOiJIUzI1NiJ9.eyJlbWFpbCI6IjI0ZHMxMDAwMTA5QGRzLnN0dWR5LmlpdG0uYWMuaW4ifQ.zxJYq62CPhPE7aKelOK7302kI9vQ3t83v00_mhfX3IU"
PROXY_URL = 'https://aiproxy.sanand.workers.dev/openai/v1/embeddings'

# Define the model and words
model_name = "text-embedding-3-small"
word1 = "Important"
word2 = "Flexible"

# Function to retrieve embedding vector for a word
def get_embedding(word):
    r_embedding = requests.post(PROXY_URL, json={
        "model": model_name,
        "input": word
    }, headers={
        "Authorization": f"Bearer {api_key}"
    })

    if r_embedding.status_code == 200:
        embedding_data = r_embedding.json()
        if 'data' in embedding_data and isinstance(embedding_data['data'], list):
            for item in embedding_data['data']:
                if 'embedding' in item:
                    return np.array(item['embedding'])
        else:
            print(f"Failed to retrieve embedding for '{word}': Unexpected response format")
    else:
        print(f"Failed to retrieve embedding for '{word}': {r_embedding.status_code}")
        print(r_embedding.text)

    return None

# Get embeddings for both words
embedding1 = get_embedding(word1)
embedding2 = get_embedding(word2)

# Calculate cosine similarity if embeddings were successfully retrieved
if embedding1 is not None and embedding2 is not None:
    similarity = 1 - cosine(embedding1, embedding2)
    print(f"Cosine similarity between '{word1}' and '{word2}': {similarity:.4f}")
```

```python
import requests
import numpy as np
from scipy.spatial.distance import cosine

api_key = "eyJhbGciOiJIUzI1NiJ9.eyJlbWFpbCI6IjI0ZHMxMDAwMTA5QGRzLnN0dWR5LmlpdG0uYWMuaW4ifQ.zxJYq62CPhPE7aKelOK7302kI9vQ3t83v00_mhfX3IU"
PROXY_URL = 'https://aiproxy.sanand.workers.dev/openai/v1/embeddings'

# Define the word lists to compare
word_lists = [
    ["Recovery", "Happiness", "Essential", "Campaign"],
    ["Majority", "Talented", "Encourage", "Industry"],
    ["Happiness", "Encourage", "Decision", "Establish"],
    ["Priority", "Attention", "Industry", "Abundance"]
]

# Function to retrieve embedding vector for a word
def get_embedding(word):
    r_embedding = requests.post(PROXY_URL, json={
        "model": "text-embedding-3-small",
        "input": word
    }, headers={
        "Authorization": f"Bearer {api_key}"
    })

    if r_embedding.status_code == 200:
        embedding_data = r_embedding.json()
        if 'data' in embedding_data and isinstance(embedding_data['data'], list):
            for item in embedding_data['data']:
                if 'embedding' in item:
                    return np.array(item['embedding'])
        else:
            print(f"Failed to retrieve embedding for '{word}': Unexpected response format")
    else:
        print(f"Failed to retrieve embedding for '{word}': {r_embedding.status_code}")
        print(r_embedding.text)

    return None

# Get embedding for the word "Optimistic"
embedding_optimistic = get_embedding("Optimistic")

# Calculate cosine similarity and average for each word list
average_similarities = []
for word_list in word_lists:
    sum_similarity = 0.0
    count = 0
    for word in word_list:
        embedding_word = get_embedding(word)
        if embedding_word is not None and embedding_optimistic is not None:
            similarity = 1 - cosine(embedding_optimistic, embedding_word)
            sum_similarity += similarity
            count += 1
    if count > 0:
        average_similarity = sum_similarity / count
        average_similarities.append(average_similarity)
    else:
        average_similarities.append(0.0)

# Find the word list with the highest average cosine similarity
max_index = np.argmax(average_similarities)
max_similarity = average_similarities[max_index]
max_word_list = word_lists[max_index]

# Print results
print(f"The word list with the highest average cosine similarity with 'Optimistic' is:")
print(max_word_list)
print(f"Average cosine similarity: {max_similarity:.4f}")
```

```python
import requests

# Define the API key and endpoint
api_key = "eyJhbGciOiJIUzI1NiJ9.eyJlbWFpbCI6IjI0ZHMxMDAwMTA5QGRzLnN0dWR5LmlpdG0uYWMuaW4ifQ.zxJYq62CPhPE7aKelOK7302kI9vQ3t83v00_mhfX3IU"
endpoint = "https://aiproxy.sanand.workers.dev/openai/v1/chat/completions"

# Define the input text
input_text = "Hej! Vad gör du här? Jag trodde att du var på semester med familjen."

# Prepare data for the request
data = {
    "model": "gpt-3.5-turbo",  # Model name
    "messages": [{"role": "system", "content": input_text}],
    "max_tokens": 1,
    "temperature": 0,
    "top_p": 1.0,
    "n": 1,
    "stop": "\n"
}

# Make the request to AI Proxy
headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json"
}

response = requests.post(endpoint, json=data, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    result = response.json()
    if 'cost' in result:
        input_cost_cents = result['cost']
        print(f"The input would cost approximately {input_cost_cents:.5f} cents.")
    else:
        print("Failed to retrieve input cost: Unexpected response format")
else:
    print(f"Failed to make request: {response.status_code}")
    print(response.text)
```
