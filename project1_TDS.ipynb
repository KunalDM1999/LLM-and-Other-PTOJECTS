```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
# Base URL
url = "https://22f3001919.github.io/tds_project_1/"

# Function to get the HTML content of a page
def get_page_content(page_url):
    response = requests.get(page_url)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to retrieve {page_url}")
        return None

# Function to extract data from the HTML content
def extract_data_from_page(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    script_tag = soup.find('script', {'type': 'module'})

    if script_tag:
        script_content = script_tag.string
    else:
        print("No script tag found")
        return None, None

    # Extract the JSON data from the script content
    try:
        data_prefix = 'const data = '
        next_prefix = 'const next = '
        data_start = script_content.find(data_prefix) + len(data_prefix)
        data_end = script_content.find(';', data_start)
        data_json = script_content[data_start:data_end].strip()

        next_start = script_content.find(next_prefix) + len(next_prefix)
        next_end = script_content.find(';', next_start)
        next_page = script_content[next_start:next_end].strip().strip('"')

        return json.loads(data_json), next_page
    except json.JSONDecodeError as e:
        print(f"JSON decoding error: {e}")
        return None, None
    except Exception as e:
        print(f"Error while extracting data: {e}")
        return None, None

# Function to get all data pages
def get_all_data_pages(start_url):
    data = []
    current_url = start_url

    while current_url:
        html_content = get_page_content(current_url)
        if html_content:
            page_data, next_page = extract_data_from_page(html_content)
            if page_data:
                data.extend(page_data)
                current_url = url + next_page if next_page else None
            else:
                break
        else:
            break

    return data

# Function to filter data by state name
def filter_data_by_state(data, state_name):
    return [item for item in data if item['ST_NAME'] == state_name]

# Main function
def main():
    start_url = url + "index.html"
    all_data = get_all_data_pages(start_url)
    himachal_data = filter_data_by_state(all_data, "HIMACHAL PRADESH")

    # Print the filtered data
    for item in himachal_data:
        print(item)

if __name__ == "__main__":
    main()
```

```python
from ast import Str
from urllib.request import urlopen
from bs4 import BeautifulSoup
import json
import pandas as pd

# Base URL and list of suffixes
base_url = 'https://22f3001919.github.io/tds_project_1/'
suffixes = ['ROnZoGPo', 'MrkYIuKb', 'iy8FeVWW', 'JRnTYVvK', 'M3gZGwCa',
            'nOOE7Dt0', 'qLRYWxdc', 'T7vBSOqP', 'swHTjbMb', 'cnOed8A3',
            'nAwku9Ux', 'zqGTQCc5', 'zyai1FTj', 'u0YpbYkW']

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each suffix
for suffix in suffixes:
    # Construct the full URL
    url = f'{base_url}{suffix}.html'

    # Fetch the content from the URL
    x = urlopen(url)

    # Parse the content using BeautifulSoup
    soup = BeautifulSoup(x.read(), 'html.parser')

    # Find the script tag and extract its content
    script_tag = soup.find('script', type='module')
    script_content = script_tag.string

    # Extract the JSON data from the script content
    data_start = script_content.find('const data = ') + len('const data = ')
    data_end = script_content.find('];', data_start) + 1
    json_data = script_content[data_start:data_end].strip()

    # Convert the JSON data to a Python list
    data = json.loads(json_data)

    # Convert JSON data to pandas DataFrame
    df = pd.DataFrame(data)

    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Display the combined DataFrame
combined_df

# Extract numeric part (integer) from column 'AC'
combined_df['AC']= combined_df['AC'].str.replace(r'^\d+\s+', '', regex=True)

# Display the modified DataFrame
combined_df
```

#***Q7***
 **For the  given AC_NAME,  In which year was the difference between the first and last candidates votes the biggest? [string]**

**If two or more years have the same difference in votes, then choose the latest year.**


```python
# Function to extract candidate rank from the candidate name
def extract_candidate_rank(candidate_name):
    try:
        # Extract the first part of the candidate name and convert to integer
        candidate_rank = int(candidate_name.split()[0])
        return candidate_rank
    except (ValueError, IndexError):
        return None  # Return None if extraction fails

# Apply the function to extract candidate ranks
project_data_1['CANDIDATE_RANK'] = project_data_1['CANDIDATE'].apply(extract_candidate_rank)

# Initialize variables to track the largest difference and its corresponding year
largest_difference = 0
year_with_largest_difference = None

# Group by year and calculate the difference between the first and last candidate votes
for year, group in project_data_1.groupby('YEAR'):
    first_candidate_votes = group.loc[group['CANDIDATE_RANK'] == 1, 'VOTES'].max()
    last_candidate_votes = group.loc[group['CANDIDATE_RANK'] == 7, 'VOTES'].min()

    if pd.notna(first_candidate_votes) and pd.notna(last_candidate_votes):
        difference = first_candidate_votes - last_candidate_votes

        # Check if this difference is larger than previously found differences
        if difference > largest_difference:
            largest_difference = difference
            year_with_largest_difference = year
        elif difference == largest_difference:
            # If two or more years have the same difference, choose the latest year
            year_with_largest_difference = max(year_with_largest_difference, year)

# Convert year_with_largest_difference to string format
year_with_largest_difference = str(year_with_largest_difference)

# Output the result
print("Year with the biggest difference between the first and last candidates' votes:", year_with_largest_difference)
```

#***Q4***

**For the  given AC_NAME, What is the highest margin percentage by which the winning candidate has defeated the runner up in any election?**

**Give your answer to 2 decimal places. [float]**


```python
import pandas as pd

# Assuming 'project_data' is already loaded into a DataFrame
# Example: project_data = pd.read_csv('your_file.csv')

# Step 1: Remove entries with NaN/NA/blank votes
valid_votes_data = project_data_1.dropna(subset=['VOTES'])

# Step 2: Calculate total votes per election year and constituency
total_votes = valid_votes_data.groupby(['YEAR', 'AC'])['VOTES'].sum().reset_index()
total_votes.columns = ['YEAR', 'AC', 'TOTAL_VOTES']

# Step 3: Merge total votes back to the original data
valid_votes_data = pd.merge(valid_votes_data, total_votes, on=['YEAR', 'AC'])

# Step 4: Calculate vote share percentage
valid_votes_data['VOTE_SHARE'] = (valid_votes_data['VOTES'] / valid_votes_data['TOTAL_VOTES']) * 100

# Step 5: Find the winner and runner-up for each election year and constituency
def calculate_margin(group):
    group = group.sort_values(by='VOTES', ascending=False)
    if len(group) > 1:
        winner_share = group.iloc[0]['VOTE_SHARE']
        runnerup_share = group.iloc[1]['VOTE_SHARE']
        margin_percentage = winner_share - runnerup_share
        return pd.Series({'MARGIN_PERCENTAGE': margin_percentage})
    else:
        return pd.Series({'MARGIN_PERCENTAGE': 0})

margin_percentages = valid_votes_data.groupby(['YEAR', 'AC']).apply(calculate_margin)

# Step 6: Find the highest margin percentage
highest_margin_percentage = margin_percentages['MARGIN_PERCENTAGE'].max()

# Round the result to 2 decimal places
highest_margin_percentage_rounded = round(highest_margin_percentage, 2)

highest_margin_percentage_rounded
```

# **Q14**
**What percentage of elections did female candidates win, when there was at least one female candidate in that election? Give your answer to 2 decimal places. [float]**



```python
import pandas as pd

# Assuming 'project_data' is already loaded into a DataFrame
# Example: project_data = pd.read_csv('your_file.csv')

# Step 1: Filter the data to include only elections with at least one female candidate
elections_with_female = project_data_1.groupby(['YEAR', 'AC']).filter(lambda x: (x['SEX'] == 'F').any())

# Step 2: Determine the winner for each election
def determine_winner(group):
    group = group.sort_values(by='VOTES', ascending=False)
    winner = group.iloc[0]
    return winner

winners = elections_with_female.groupby(['YEAR', 'AC']).apply(determine_winner)

# Step 3: Calculate the percentage of elections won by female candidates
total_elections_with_female = winners.shape[0]
female_winners = winners[winners['SEX'] == 'F'].shape[0]

percentage_female_winners = (female_winners / total_elections_with_female) * 100

# Round the result to 2 decimal places
percentage_female_winners_rounded = round(percentage_female_winners, 2)

percentage_female_winners_rounded
```

**Find the Pearson correlation coefficient between the votes won by female candidates and male candidates. Only include constituencies and election years that had at least 1 female candidate. Each row in the correlation data table should represent an election. Round off the answer to 4 decimal points[float]**

```python
# Filter rows where SEX is 'F'
female_candidates_df = combined_df[combined_df['SEX'] == 'F']

# Group by YEAR and AC to find constituencies and years with at least one female candidate
grouped_female_candidates = female_candidates_df.groupby(['YEAR', 'AC']).size().reset_index(name='count')

# Filter the original DataFrame to include only these constituencies and years
filtered_df = combined_df.merge(grouped_female_candidates[['YEAR', 'AC']], on=['YEAR', 'AC'])

# Initialize an empty dictionary to store correlation matrices
correlation_matrices = {}

# Compute correlation matrix for each election year
for year in filtered_df['YEAR'].unique():
    election_data = filtered_df[filtered_df['YEAR'] == year]
    # Select only numeric columns for correlation calculation
    numeric_data = election_data.select_dtypes(include=[float, int])
    correlation_matrix = numeric_data.corr()
    correlation_matrices[year] = correlation_matrix

# Display the correlation matrices
for year, corr_matrix in correlation_matrices.items():
    print(f"Correlation matrix for the year {year}:\n{corr_matrix}\n")
```

# **Q18**
**For a given state, identify the election year which has the most outliers on the basis of candidate votes. Use the interquartile range (IQR) rule to detect outliers. [integer]**

```python
import pandas as pd

outlier_counts = {}
for year in combined_df['YEAR'].unique():
    # Get the votes for the current year
    year_votes = combined_df[combined_df['YEAR'] == year]['VOTES'].dropna()

    # Calculate Q1, Q3, and IQR
    Q1 = year_votes.quantile(0.25)
    Q3 = year_votes.quantile(0.75)
    IQR = Q3 - Q1

    # Define the bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = year_votes[(year_votes < lower_bound) | (year_votes > upper_bound)]

    # Store the count of outliers for the current year
    outlier_counts[year] = len(outliers)

# Identify the year with the most outliers
year_with_most_outliers = max(outlier_counts, key=outlier_counts.get)

year_with_most_outliers, outlier_counts[year_with_most_outliers]
```
