```python
import pandas as pd

# Assuming your DataFrame is named df

# Group by 'tag_id' and count the number of unique 'goodreads_book_id' for each tag
tag_book_counts = df_tags.groupby('tag_id')['goodreads_book_id'].nunique()

# Find the maximum number of books associated with any tag
max_books_count = tag_book_counts.max()

# Filter the tags that have this maximum count
most_frequent_tags = tag_book_counts[tag_book_counts == max_books_count]

# Get the tag_id with the least numerical value among the most frequent tags
most_frequent_tag = most_frequent_tags.idxmin()

print(f"The most frequently used tag_id is: {most_frequent_tag}")
```

```python
df_books_1 = df_books[df_books['original_title'].notnull()]
df_books_1.info()
```

```python
from textblob import TextBlob
# Create a function to calculate the sentiment polarity
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity
# Apply the function to 'original_title' and create a new column 'sentiment'
df_books_1['sentiment'] = df_books_1['original_title'].apply(get_sentiment)

# Classify titles as positive (polarity >= 0)
df_positive_titles = df_books_1[df_books_1['sentiment'] >= 0]

# Count the number of positive titles
positive_titles_count = df_positive_titles.shape[0]

# Output the result
print(f"Number of positive titles: {positive_titles_count}")
```

```python
from textblob import TextBlob

# Calculate sentiment polarity for each title
df_books['sentiment'] = df_books['original_title'].apply(lambda title: TextBlob(title).sentiment.polarity)

# Count positive titles
num_positive = (df_books['sentiment'] >= 0).sum()

print("Number of positive titles:", num_positive)
```
